```python
import matplotlib.pyplot as plt
import seaborn

import pandas as pd
import numpy as np

# 训练数据
data_train=pd.read_csv("C:\\Users\\yuanf\\introdution_data_python\\week10\\train.csv")
# 测试数据
data_test = pd.read_csv("C:\\Users\\yuanf\\introdution_data_python\\week10\\test.csv")

# 显示中文标题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 查看有多少人丧生，多少人获救
print(data_train['Survived'].value_counts())

 # 绘图
fig = plt.figure()
# 乘客等级分布
plt.subplot2grid((2, 3), (0, 0))
data_train['Pclass'].value_counts().plot(kind='bar')
plt.ylabel(u"人数")
plt.xlabel(u'乘客等级')
plt.title(u'乘客等级分布')

# 乘客性别分布
plt.subplot2grid((2, 3), (0, 1))
data_train['Sex'].value_counts().plot(kind='bar')
plt.ylabel(u"人数")
plt.xlabel(u'性别')
plt.title(u'乘客性别分布')

# 乘客的年龄分布
plt.subplot2grid((2, 3), (0, 2))
data_train['Age'].hist()
plt.xlabel(u'年龄')
plt.title(u'乘客年龄分布')

plt.show()
```
