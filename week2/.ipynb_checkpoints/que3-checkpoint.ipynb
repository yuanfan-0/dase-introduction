```python
#建立一个含四个元素的list来表示人羊狼菜的状态如[1,1,1,1]表示4个元素都在
left_forbid_states = [[0,1,1,1],[0,1,1,0],[0,1,0,1],[0,0,0,0]]
right_forbid_states = [[0,1,1,1],[0,1,1,0],[0,1,0,1],[1,1,1,1]]
init_state = [0,0,0,0]
end_state = [1,1,1,1]
#发现狼和菜两个元素在条件判断索取的条件是相同的，因此有个对称过程
actions1=[]
actions2=[]
elements1=["人","羊","狼","菜"]
elements2=["人","羊","菜","狼"]
#定义状态判断函数
def condition_judge(state,index):
    if index == 1:
        for i in right_forbid_states :
            if (state == i):
                return False
    else :
        for i in left_forbid_states :
            if (state == i):
                return False
    return True
#定义一次人划船过程
def rowing(left_state,right_state):
    
    if right_state[0] == 1:
        for i in range(1,4):
            temp_left = [i for i in left_state]
            temp_right = [i for i in right_state]
            if right_state[i] == 1:
                temp_left[0] = 1
                temp_left[i] = 1
                temp_right[0] = 0
                temp_right[i] = 0
                
                if condition_judge(temp_left,2) and condition_judge(temp_right,1):
                    for j in range(4):
                        left_state[j] = temp_left[j]
                        right_state[j] = temp_right[j]
                    cur_left = [0,0,0,0]
                    cur_right = [0,0,0,0]
                    for k in range(4):
                        cur_left[k] = temp_left[k] 
                        cur_right[k] = temp_right[k]
                    left_forbid_states.append(cur_left)
                    right_forbid_states.append(cur_right)
                    
                    action1 = "人带%s从右岸到左岸"%elements1[i]
                    action2= "人带%s从右岸到左岸"%elements2[i]
                    actions1.append(action1)
                    actions2.append(action2)
                    return
                
        temp_left = [i for i in left_state]
        temp_right = [i for i in right_state]
        temp_left[0] = 1
        temp_right[0] = 0
        if condition_judge(temp_left,2) and condition_judge(temp_right,1):
            for j in range(4):
                left_state[j] = temp_left[j]
                right_state[j] = temp_right[j]
            cur_left = [0,0,0,0]
            cur_right = [0,0,0,0]
            for k in range(4):
                cur_left[k] = temp_left[k] 
                cur_right[k] = temp_right[k]
            left_forbid_states.append(cur_left)
            right_forbid_states.append(cur_right)
            action = "人从右岸到左岸"
            actions1.append(action)
            actions2.append(action)
            return
        
    else:
        temp_left = [i for i in left_state]
        temp_right = [i for i in right_state]
        temp_left[0] = 0
        temp_right[0] = 1
       
        if condition_judge(temp_left,2) and condition_judge(temp_right,1):
            for j in range(4):
                left_state[j] = temp_left[j]
                right_state[j] = temp_right[j]
            cur_left = [0,0,0,0]
            cur_right = [0,0,0,0]
            for k in range(4):
                cur_left[k] = temp_left[k] 
                cur_right[k] = temp_right[k]

            left_forbid_states.append(cur_left)
            right_forbid_states.append(cur_right)
            action = "人从右岸到左岸"
            actions1.append(action)
            actions2.append(action)
            return
        for i in range(1,4):
            temp_left = [i for i in left_state]
            temp_right = [i for i in right_state]
            if left_state[i] == 1:
                temp_left[0] = 0
                temp_left[i] = 0
                temp_right[0] = 1
                temp_right[i] = 1
                if condition_judge(temp_left,2) and condition_judge(temp_right,1):
                    for j in range(4):
                        left_state[j] = temp_left[j]
                        right_state[j] = temp_right[j]
                    cur_left = [0,0,0,0]
                    cur_right = [0,0,0,0]
                    for k in range(4):
                        cur_left[k] = temp_left[k] 
                        cur_right[k] = temp_right[k]
                    left_forbid_states.append(cur_left)
                    right_forbid_states.append(cur_right)
                    action1 = "人带%s从右岸到左岸"%elements1[i]
                    action2= "人带%s从右岸到左岸"%elements2[i]
                    actions1.append(action1)
                    actions2.append(action2)
                    return
                            
                
        
#定义结果打印函数         
def print_actions():
    print("第一种渡河方案:")
    for i in actions1:
        print(i)
    print("\n")
    print("第二种渡河方案:")
    for j in actions2:
        print(j)
            
#定义渡河过程               
def crossing():
    left_state = [i for i in init_state]
    right_state = [i for i in end_state]
    count = 0
    while not((left_state==end_state)):
        rowing(left_state,right_state)
 
    print_actions()

crossing()
```
