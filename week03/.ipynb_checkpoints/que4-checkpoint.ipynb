```python
class Node:
    
     #节点初始化
    def __init__(self, value = None, next = None):
        self.value = value
        self.next = next
class Linklist:
    
    #链表初始化
    def __init__(self):
        self.head = Node()
        self.size = 0
    
    #头插法
    def insert_head(self,value):
        node = Node(value)
        node.next = self.head.next
        self.head.next = node
        self.size += 1
    
    #尾插法
    def insert_tail(self,value):
        node = Node(value)
        temp = self.head
        while temp.next != None:
            temp = temp.next
        temp.next = node
        self.size += 1
        
    #删除指定节点
    def delete_node(self,target):
        temp = self.head
        index = 0
        while temp.next != None:
            pre = temp
            temp = temp.next
            if temp.value == target:
                index = 1
                pre.next = temp.next
                self.size -= 1
        if not index:
            print("链表中未找到符合该元素的节点")
            
    #查找操作
    def find_node(self,target):
        temp = self.head
        count = 0
        while temp.next != None:
            temp = temp.next
            count += 1
            if temp.value == target:
                print("该元素在第%d个节点!"%count)
    #修改操作
    def amend_node(self,pos,target):  #将第pos个位置节点的值修改为target
        temp = self.head
        for i in range(0,pos):
            temp = temp.next
        temp.value = target
    
    #链表打印功能
    def print_linklist(self):
        print("链表内容：")
        temp = self.head
        for i in range(0,self.size-1):
            temp = temp.next
            print("%d----->"%temp.value,end='')
        print(temp.next.value)


index = input("请输入操作(0代表退出，1代表头插入，2代表尾插入，3代表删除，4代表查找，5代表修改，6代表打印链表内容):")

mylinklist = Linklist()
while index != '0':
    if index == '1':
        elem = int(input("请输入头插入值:"))
        mylinklist.insert_head(elem)
    elif index == '2':
        elem = int(input("请输入尾插入值:"))
        mylinklist.insert_tail(elem)
    elif index == '3':
        target = int(input("请输入要删除的值:"))
        mylinklist.delete_node(target)
    elif index == '4':
        target = int(input("请输入要查找的值:"))
        mylinklist.find_node(target)
    elif index == '5':
        target = input("请输入要修改的节点与值(pos target):")
        a = target.split(" ")
        pos = int(a[0])
        target = int(a[1])
        mylinklist.amend_node(pos,target)
    else:
        mylinklist.print_linklist()
    index = input("请输入操作(0代表退出，1代表头插入，2代表尾插入，3代表删除，4代表查找，5代表修改，6代表打印链表内容):")

        
                
```
