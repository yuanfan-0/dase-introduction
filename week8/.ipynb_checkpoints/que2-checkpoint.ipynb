### ①打印数据集


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data_load=sns.load_dataset('car_crashes')
data_load.sample(10)

```

### ②柱状图

```python
plt.rcParams['font.sans-serif']=['SimHei']
# 创建直方图
x = [i for i in range(0,51)]
plt.bar(x,data_load['speeding'])
plt.title('柱状图')
plt.xlabel('id_num')
plt.ylabel('speed')
plt.show()
```

### ③散点图

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.scatter(x,data_load['speeding'],c='red', marker='o', label='Speed Data')
plt.title('Scatter Plot')
plt.xlabel('id_num')
plt.ylabel('speed')
plt.legend()
plt.show()
```

### ④饼图

```python
plt.pie(data_load[40:]['ins_losses'],labels = data_load[40:]['abbrev'])
plt.title('pie graph')
plt.show()
```
