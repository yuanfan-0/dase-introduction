### ①查看数据

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data_load=sns.load_dataset('car_crashes')
data_load.sample(10)
```

### ②直方图

```python

data_speeding = data_load['speeding']
sns.histplot(data_speeding,kde=True)
```

### ③分类散点图

```python
sns.swarmplot(x="abbrev", y="ins_premium", hue="ins_losses", data=data_load[40:])
```

### ④点图

```python
sns.pointplot(x="abbrev", y="ins_premium", data=data_load[30:])
```
